```python
#creating pivot table
book_pivot=pd.pivot_table(data=books,values='Book.Rating',index='User.ID',columns='Book.Title')
book_pivot
```

```python
book_pivot.isna().sum()
```

```python
#filling NaN values as zeros
book_pivot.fillna(0,inplace=True)
book_pivot.head()
```

```python
#model building
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import correlation,cosine
```

```python
rating_data=1-pairwise_distances(X=book_pivot,metric='cosine')
rating_data
```

```python
rating_df=pd.DataFrame(rating_data)
rating_df
```

```python
rating_df.index=books['User.ID'].unique()
rating_df.columns=books['User.ID'].unique()
```

```python
users_50=rating_df.iloc[:50,:50]
users_50
```
