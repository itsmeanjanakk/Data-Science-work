```python
#drop unnecessory columns
data=forestfire.copy()
```

```python
data.columns
```

```python
data.drop(columns=['month','day'],axis=1,inplace=True)
```

```python
data.columns
```

```python
#conveting to numeric values
from sklearn.preprocessing import LabelEncoder
```

```python
label=LabelEncoder()
```

```python
data['size_category']=label.fit_transform(data['size_category'])
```

```python
data.head()
```

```python
#split into input x and output y variables
x=data.iloc[:,0:9]
```

```python
x
```

```python
y=data[['size_category']]
```

```python
y
```

```python
x.corr()
```

```python
from tensorflow.keras import datasets,layers,models
```

```python
#model building
seq_model=models.Sequential([
    layers.Dense(16,input_dim = 9, activation = "relu"),
    layers.Dense(32, activation = "relu"),
    layers.Dense(1,activation="sigmoid")
])
```

```python
seq_model.compile(optimizer="adam",loss="binary_crossentropy",metrics="accuracy")
```

```python
seq_model.fit(x,y,epochs=400)
```

```python
seq_model.evaluate(x,y)
```

```python
prd=seq_model.predict(x)
prd_df=pd.DataFrame(prd)
prd_df.head(20)
```
