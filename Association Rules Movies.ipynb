```python
movies_data=movies.copy()
```

```python
movies_data.head()
```

```python
movies_data.drop(columns=['V1','V2','V3','V4','V5'],axis=1,inplace=True)
```

```python
movies_data.columns
```

```python
movies_data.shape
```

```python
frequent_items=apriori(df=movies_data,min_support=0.5,use_colnames=True)
```

```python
frequent_items
```

```python
rules=association_rules(df=frequent_items,metric='confidence',min_threshold=0.8)
rules
```
